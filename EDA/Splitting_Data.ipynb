# The goal is to join the two datasets, so that we can use the modality column in the second dataset to split the dataset into the different modalities

**Feature engineering to ensure that the key to join the dataset is identical**

```python
data['report_id'] = data['report_id'].astype(str)

# Create the new 'join' column by removing 'RAD_'
data['join'] = data['report_id'].str.replace('RAD_', '', regex=False)

# Convert the 'join' column back to integer
data['join'] = pd.to_numeric(data['join'], errors='coerce')

data.head()
```

```python
data.dtypes
```

# Join Dataset

```python
%%time
# Perform the merge
merged_data = pd.merge(data, df[['report_id', 'modality']], left_on='join', right_on='report_id', how='left')

# Remove the redundant 'report_id' column from the result
merged_data = merged_data.drop('report_id', axis=1)

```

```python
merged_data
```

```python
# Display info about the resulting DataFrame
print(merged_data.info())
```

# Create instruction column for instruction-tuning

```python
%%time
def set_instruction(modality):
    if pd.isna(modality):
        return None
    elif modality.upper() == 'US':
        return "Convert the findings and clinical information of the ultrasound to an impression"
    elif modality.upper() == 'XRAY':
        return "Convert the findings and clinical information of the X-Ray to an impression"
    elif modality.upper() == 'FL':
        return "Convert the findings and clinical information of the fluids to an impression"

    else:
        return f"Convert the findings and clinical information of the {modality} result to an impression"

# Add the new 'instruction' column
merged_data['instruction'] = merged_data['modality'].apply(set_instruction)

# Display value counts to see the distribution of instructions
print(merged_data['instruction'].value_counts(dropna=False))
```

```python
merged_data.info()
```

# Seperate dataset by modality


```python
%%time
# Get unique modality values
modalities = merged_data['modality'].unique()

# Create a dictionary to store the split DataFrames
modality_dfs = {}

# Split the DataFrame for each modality
for modality in modalities:
    if pd.isna(modality):
        modality_dfs['Unknown'] = merged_data[merged_data['modality'].isna()]
    else:
        modality_dfs[modality] = merged_data[merged_data['modality'] == modality]

# Print information about each split DataFrame
for modality, df in modality_dfs.items():
    print(f"\nModality: {modality}")
    print(f"Number of rows: {len(df)}")
    print(df.head(2))  # Print first two rows of each DataFrame

```

```python
ct_df = modality_dfs['CT']
ct_df.info()
```

**Create a small sample dataset for inital testing of training scripts**

```python
# Create a dictionary to store random sample of split DataFrames
modality_sample = {}

# Create 50 random samples for each modality
for modality, df in modality_dfs.items():
    # Filter out rows where clinical_information, findings, or impression are null
    filtered_df = df.dropna(subset=['clinical_information', 'findings', 'impression'])

    # Ensure there are at least 50 rows for sampling
    if len(filtered_df) >= 50:
        modality_sample[modality] = filtered_df.sample(n=50, random_state=42)
    else:
        print(f"Warning: Modality '{modality}' has less than 50 valid rows. Using the entire filtered DataFrame.")
        modality_sample[modality] = filtered_df  # Use all available rows

# Print information about each split DataFrame
for modality, df in modality_sample.items():
    print(f"\nModality: {modality}")
    print(f"Number of rows: {len(df)}")
    print(df.head(2))  # Print first two rows of each DataFrame
```
